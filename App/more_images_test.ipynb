```python
import cv2
import tensorflow as tf
from keras.preprocessing import image
import numpy as np  

Categories = ["Closed","Left","Right"]
#Categories = ["Closed","Left","Open","Right"]

def prepare(filepath):
    img_size = 224
    img_array = cv2.imread(filepath)
    new_array = cv2.resize(img_array,(img_size,img_size),3)
    return new_array.reshape(-1,img_size,img_size,3)

model = tf.keras.models.load_model(r"---testte kullanılacak olan modelin yolu---")
rightSayac = 0
leftSayac = 0
closedSayac = 0
openSayac = 0


for i in range(1,90):
    #resmi tek bir formatta yazılırsa çalışır. örnek: resim (1), resim (2)
    #bunun için tüm resimleri seçik f2 tuşuna basıp "resim" ismini verdiğinde hepsini
    #otomatik olarak düzenliyor.
    prediction = model.predict([prepare(r'---resmin yolu ('+str(i)+').jpg')])
    indis = np.argmax(prediction[0])
    
    img = image.load_img(r'C:\MyProject\397199_397209\Data\ThreeClassDataBase\Test\Closed\Closed ('+str(i)+').jpg')

    print("TEST SONUCU "+str(i)+" :"+Categories[indis])

    if Categories[indis] == "Closed":
        closedSayac+=1
    elif Categories[indis] == "Left":
        leftSayac+=1
    #elif Categories[indis] == "Open": openSayac+=1
    else:
        rightSayac+=1
    
print("Closed Sonucu: ",closedSayac)
print("Left Sonucu: ",leftSayac)
#print("Open Sonucu: ",openSayac)
print("Right Sonucu: ",rightSayac)
```
